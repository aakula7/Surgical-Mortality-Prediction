```python
def changetype(elem):
    try:
        return int(elem)
    except:
        return 0
    
df.PRINC_DIAG_CODE = df['PRINC_DIAG_CODE'].map(lambda x: changetype(x))
df.PRINC_ICD9_CODE = df['PRINC_ICD9_CODE'].map(lambda x: changetype(x))
df.PRINC_SURG_PROC_CODE = df['PRINC_SURG_PROC_CODE'].map(lambda x: changetype(x))
```
