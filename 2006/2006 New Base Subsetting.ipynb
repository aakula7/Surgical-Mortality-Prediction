```python
import pandas as pd
```

```python
Base06 = pd.read_csv('Base_06_Subset.csv', index_col = 0)
Base06.head()
```

```python
columnsKeep = ['RECORD_ID', 'PRINC_DIAG_CODE', 'OTH_DIAG_CODE_1', 'OTH_DIAG_CODE_2', 'OTH_DIAG_CODE_3', 'OTH_DIAG_CODE_4', 'OTH_DIAG_CODE_5', 'OTH_DIAG_CODE_6', 'OTH_DIAG_CODE_7', 'OTH_DIAG_CODE_8', 'OTH_DIAG_CODE_9', 'OTH_DIAG_CODE_10', 'OTH_DIAG_CODE_11', 'OTH_DIAG_CODE_12', 'OTH_DIAG_CODE_13', 'OTH_DIAG_CODE_14', 'OTH_DIAG_CODE_15', 'OTH_DIAG_CODE_16', 'OTH_DIAG_CODE_17', 'OTH_DIAG_CODE_18', 'OTH_DIAG_CODE_19', 'OTH_DIAG_CODE_20', 'OTH_DIAG_CODE_21', 'OTH_DIAG_CODE_22', 'OTH_DIAG_CODE_23', 'OTH_DIAG_CODE_24']
newBase06 = Base06[columnsKeep]
newBase06.head()
```

```python
newBase06 = newBase06.fillna(0)
newBase06.isnull().sum()
```

```python
newBase06.head()
```

```python
NB06_Dict = newBase06.to_dict('RECORD_ID')
NB06_Dict
```

```python
ID = []
Diag_Code = []

for dic1 in NB06_Dict:
    ptID = dic1['RECORD_ID']
    for dic2 in dic1:
        if dic2 != 'RECORD_ID':
            ID.append(ptID)
            Diag_Code.append(dic1[dic2])
```

```python
CodesDF = pd.DataFrame({'RECORD_ID':ID,
                       'CODE':Diag_Code})
CodesDF
```

```python
CodesDF = CodesDF[CodesDF.CODE != 0]
CodesDF
```

```python
CodesDF.to_csv('Base_06_Codes.csv')
```

```python
def oneHot(value):
    expList = [20, 40, 41, 42]
    if value in expList:
        return 1
    else:
        return 0

Base06.PAT_STATUS = Base06['PAT_STATUS'].map(lambda x: oneHot(x))
Base06.PAT_STATUS
```
