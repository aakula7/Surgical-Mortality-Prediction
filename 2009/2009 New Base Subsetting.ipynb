```python
import pandas as pd
```

```python
Base09 = pd.read_csv('Base_09_Subset.csv', index_col = 0)
Base09.head()
```

```python
columnsKeep = ['RECORD_ID', 'PRINC_DIAG_CODE', 'OTH_DIAG_CODE_1', 'OTH_DIAG_CODE_2', 'OTH_DIAG_CODE_3', 'OTH_DIAG_CODE_4', 'OTH_DIAG_CODE_5', 'OTH_DIAG_CODE_6', 'OTH_DIAG_CODE_7', 'OTH_DIAG_CODE_8', 'OTH_DIAG_CODE_9', 'OTH_DIAG_CODE_10', 'OTH_DIAG_CODE_11', 'OTH_DIAG_CODE_12', 'OTH_DIAG_CODE_13', 'OTH_DIAG_CODE_14', 'OTH_DIAG_CODE_15', 'OTH_DIAG_CODE_16', 'OTH_DIAG_CODE_17', 'OTH_DIAG_CODE_18', 'OTH_DIAG_CODE_19', 'OTH_DIAG_CODE_20', 'OTH_DIAG_CODE_21', 'OTH_DIAG_CODE_22', 'OTH_DIAG_CODE_23', 'OTH_DIAG_CODE_24']
newBase09 = Base09[columnsKeep]
newBase09.head()
```

```python
newBase09 = newBase09.fillna(0)
newBase09.isnull().sum()
```

```python
newBase09.head()
```

```python
NB09_Dict = newBase09.to_dict('RECORD_ID')
NB09_Dict
```

```python
ID = []
Diag_Code = []

for dic1 in NB09_Dict:
    ptID = dic1['RECORD_ID']
    for dic2 in dic1:
        if dic2 != 'RECORD_ID':
            ID.append(ptID)
            Diag_Code.append(dic1[dic2])
```

```python
CodesDF = pd.DataFrame({'RECORD_ID':ID,
                       'CODE':Diag_Code})
CodesDF
```

```python
CodesDF = CodesDF[CodesDF.CODE != 0]
CodesDF
```

```python
CodesDF.to_csv('Base_09_Codes.csv')
```

```python
def oneHot(value):
    expList = [20, 40, 41, 42]
    if value in expList:
        return 1
    else:
        return 0

Base09.PAT_STATUS = Base09['PAT_STATUS'].map(lambda x: oneHot(x))
Base09.PAT_STATUS
```
