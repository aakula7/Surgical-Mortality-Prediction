```python
import pandas as pd
```

```python
Base10 = pd.read_csv('Base_10_Subset.csv', index_col = 0)
Base10.head()
```

```python
columnsKeep = ['RECORD_ID', 'PRINC_DIAG_CODE', 'OTH_DIAG_CODE_1', 'OTH_DIAG_CODE_2', 'OTH_DIAG_CODE_3', 'OTH_DIAG_CODE_4', 'OTH_DIAG_CODE_5', 'OTH_DIAG_CODE_6', 'OTH_DIAG_CODE_7', 'OTH_DIAG_CODE_8', 'OTH_DIAG_CODE_9', 'OTH_DIAG_CODE_10', 'OTH_DIAG_CODE_11', 'OTH_DIAG_CODE_12', 'OTH_DIAG_CODE_13', 'OTH_DIAG_CODE_14', 'OTH_DIAG_CODE_15', 'OTH_DIAG_CODE_16', 'OTH_DIAG_CODE_17', 'OTH_DIAG_CODE_18', 'OTH_DIAG_CODE_19', 'OTH_DIAG_CODE_20', 'OTH_DIAG_CODE_21', 'OTH_DIAG_CODE_22', 'OTH_DIAG_CODE_23', 'OTH_DIAG_CODE_24']
newBase10 = Base10[columnsKeep]
newBase10.head()
```

```python
newBase10 = newBase10.fillna(0)
newBase10.isnull().sum()
```

```python
newBase10.head()
```

```python
NB10_Dict = newBase10.to_dict('RECORD_ID')
NB10_Dict
```

```python
ID = []
Diag_Code = []

for dic1 in NB10_Dict:
    ptID = dic1['RECORD_ID']
    for dic2 in dic1:
        if dic2 != 'RECORD_ID':
            ID.append(ptID)
            Diag_Code.append(dic1[dic2])
```

```python
CodesDF = pd.DataFrame({'RECORD_ID':ID,
                       'CODE':Diag_Code})
CodesDF
```

```python
CodesDF = CodesDF[CodesDF.CODE != 0]
CodesDF
```

```python
CodesDF.to_csv('Base_10_Codes.csv')
```

```python
Charlson10 = pd.read_csv('Charlson_10.csv', index_col = 0)
Charlson10.head()
```

```python
Charlson10 = Charlson10.rename(columns = {'score':'CHARLSON_SCORE',
                                         'index':'CHARLSON_INDEX',
                                         'wscore':'CHARLSON_WSCORE',
                                         'windex':'CHARLSON_WINDEX'})
Charlson10.head()
```

```python
Base10 = pd.merge(Base10, Charlson10, how = 'outer', on = 'RECORD_ID')
Base10.head()
```

```python
Elixhauser10 = pd.read_csv('Elixhauser_10.csv', index_col = 0)
Elixhauser10.head()
```

```python
Elixhauser10 = Elixhauser10.rename(columns = {'score':'ELIXHAUSER_SCORE',
                                             'index':'ELIXHAUSER_INDEX',
                                             'wscore_ahrq':'ELIXHAUSER_WSCORE_AHRQ',
                                             'wscore_vw':'ELIXHAUSER_WSCORE_VW',
                                             'windex_ahrq':'ELIXHAUSER_WINDEX_AHRQ',
                                             'windex_vw':'ELIXHAUSER_WINDEX_VW'})
Elixhauser10.head()
```

```python
Base10 = pd.merge(Base10, Elixhauser10, how = 'outer', on = 'RECORD_ID')
Base10.head()
```

```python
def oneHot(value):
    expList = [20, 40, 41, 42]
    if value in expList:
        return 1
    else:
        return 0

Base10['PAT_STATUS_CLASS'] = Base10['PAT_STATUS'].map(lambda x: oneHot(x))
Base10['PAT_STATUS_CLASS']
```

```python
Base10 = Base10[['RECORD_ID', 'THCIC_ID', 'SPEC_UNIT', 'SEX_CODE', 'TYPE_OF_ADMISSION', 'SOURCE_OF_ADMISSION',
                 'ADMIT_WEEKDAY', 'LENGTH_OF_STAY', 'PAT_AGE', 'PAT_STATUS', 'PAT_STATUS_CLASS', 'RACE', 'ETHNICITY',
                 'ADMITTING_DIAGNOSIS', 'PRINC_DIAG_CODE', 'PRINC_SURG_PROC_CODE', 'PRINC_ICD9_CODE', 'RISK_MORTALITY', 
                 'ILLNESS_SEVERITY', 'CHARLSON_SCORE', 'CHARLSON_INDEX', 'CHARLSON_WSCORE', 'CHARLSON_WINDEX', 
                 'ELIXHAUSER_SCORE', 'ELIXHAUSER_INDEX', 'ELIXHAUSER_WSCORE_AHRQ', 'ELIXHAUSER_WSCORE_VW',
                 'ELIXHAUSER_WINDEX_AHRQ', 'ELIXHAUSER_WINDEX_VW']]
Base10.head()
```
